```python
import os
import pandas as pd

# ── Dataset paths (Kaggle mounts datasets here automatically) ─────────────────
BASE_DIR    = "/kaggle/input/isic-2019"
CSV_PATH    = "/kaggle/input/datasets/andrewmvd/isic-2019/ISIC_2019_Training_GroundTruth.csv"
IMAGE_DIR   = "/kaggle/input/datasets/andrewmvd/isic-2019/ISIC_2019_Training_Input/ISIC_2019_Training_Input"

# ── 1. Load metadata CSV ──────────────────────────────────────────────────────
df = pd.read_csv(CSV_PATH)
print("Columns:", df.columns.tolist())
print()

# ── 2. Total number of images ─────────────────────────────────────────────────
print(f"Total images (from CSV): {len(df)}")
print()

# ── 3. Available classes ──────────────────────────────────────────────────────
class_columns = [c for c in df.columns if c != "image"]
print(f"Available classes ({len(class_columns)}): {class_columns}")
print()

# ── 4. Number of images per class ─────────────────────────────────────────────
class_counts = df[class_columns].sum().astype(int).sort_values(ascending=False)
print("Images per class:")
print(class_counts.to_string())
print()

# ── 5. Cross-check with actual files on disk ──────────────────────────────────
image_files = [f for f in os.listdir(IMAGE_DIR)
               if f.lower().endswith((".jpg", ".jpeg", ".png"))]
print(f"Image files found on disk: {len(image_files)}")
```

```python
import os
from collections import defaultdict

# ── Exact base path ───────────────────────────────────────────────────────────
BASE_DIR = "/kaggle/input/datasets/orvile/brain-cancer-mri-dataset/Brain_Cancer raw MRI data/Brain_Cancer"

IMAGE_EXTENSIONS = (".jpg", ".jpeg", ".png", ".bmp", ".tiff", ".tif", ".webp")

# ── Count images per class folder ─────────────────────────────────────────────
class_counts = {}

for class_name in sorted(os.listdir(BASE_DIR)):
    class_dir = os.path.join(BASE_DIR, class_name)
    if os.path.isdir(class_dir):
        count = sum(1 for f in os.listdir(class_dir)
                    if f.lower().endswith(IMAGE_EXTENSIONS))
        class_counts[class_name] = count

# ── Results ───────────────────────────────────────────────────────────────────
print(f"Available Classes ({len(class_counts)}): {list(class_counts.keys())}")
print()
print("=== Images Per Class ===")
for cls, count in sorted(class_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"  {cls:<30} : {count}")
print()
print(f"=== TOTAL images: {sum(class_counts.values())} ===")
```
